### Use of force visualisations

```python
#heatmap of day_of_week vs hour_of_day to see peak times for use-of-force incidents
query = '''
SELECT day_of_week, hour_of_day FROM inference_view
WHERE Date_time BETWEEN '2019-01-01 00:00:00' AND '2022-03-31 23:59:59' AND
max_tactic != 0 AND
main_duty NOT IN ('Off duty', 'Custody') AND
primary_conduct > 0;
'''
df = pd.read_sql_query(query, db_connection)
#map the mysql day of week numbers to names
day_of_week_mapping = {
    1: 'Sunday',
    2: 'Monday',
    3: 'Tuesday',
    4: 'Wednesday',
    5: 'Thursday',
    6: 'Friday',
    7: 'Saturday'
}
df['day_of_week'] = df['day_of_week'].map(day_of_week_mapping)
#pivot the table to get counts of incidents by day of week and hour of day
df = df.pivot_table(index='day_of_week', 
                                         columns='hour_of_day', 
                                         aggfunc='size', 
                                         fill_value=0)
#reorder the days of the week
days_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
df = df.reindex(days_order)
#set the font size
plt.rcParams['font.size'] = 8
#plot the heatmap
plt.figure(figsize=(12, 6), dpi=200)
sns.heatmap(df, cmap='YlGnBu', annot=True, fmt='d')
plt.title('Non-compliant subject use of force incidents by day of week and hour')
plt.xlabel('Hour of Day')
plt.ylabel('Day of Week')
plt.savefig('plots/pa_heatmap_day_hour.png', dpi=200, bbox_inches='tight')
plt.show()
```

```python
query = '''
SELECT * FROM inference_view
WHERE Date_time BETWEEN '2021-04-01 00:00:00' AND '2022-03-31 23:59:59' AND
max_tactic != 0 AND
main_duty NOT IN ('Off duty', 'Custody') AND
primary_conduct > 0;
'''
df = pd.read_sql_query(query, db_connection)

#adjust the location_cols if they have been redefined elsewhere
location_cols = ['location_street', 'location_pubtrans', 'location_retail', 
                 'location_openground', 'location_licensed', 'location_hospital', 
                 'location_mentalhealth', 'location_dwelling', 'location_other']

#filter out rows where primary_conduct is 0
df = df[df['primary_conduct'] != 0]
#melting the dataframe to get a binary column for each location
melted_df = pd.melt(df, id_vars=['primary_conduct'], value_vars=location_cols, var_name='location', value_name='is_present')
#filter rows where location is not present
melted_df = melted_df[melted_df['is_present'] == 1]
#extract the label from a location column
melted_df['location_label'] = melted_df['location'].str.split('_').str[1]
#pivot the data
pivot_df = melted_df.groupby(['location_label', 'primary_conduct']).size().unstack().fillna(0)
#sort the locations by the total number of incidents occurred there
pivot_df['total'] = pivot_df.sum(axis=1)
pivot_df = pivot_df.sort_values(by='total', ascending=False)
pivot_df.drop(columns=['total'], inplace=True)
#get the percentage of incidents for each location
pivot_df = pivot_df.divide(pivot_df.sum(axis=1), axis=0) * 100

#create a color map for the stacked bar chart 
colors = ['#FFC300', #yellow
          '#FF5733', #orange
          '#C70039', #light red
          '#900C3F', #dark red
          '#581845'  #purple 
          ]
#set the font size
plt.rcParams['font.size'] = 12
#plot the stacked bar chart
ax = pivot_df.plot(kind='bar', stacked=True, figsize=(15, 7), color=colors)

ax.set_xlabel('Location')
ax.set_ylabel('Share of incidents (%)')
ax.set_title('Breakdown of non-compliant subject incidents by location and primary conduct')
ax.set_xticklabels(pivot_df.index, rotation=45)

#adding the labels with percents to the bars
for bar in ax.patches:
    width = bar.get_width()
    height = bar.get_height()
    x, y = bar.get_xy()
    if height > 0:
        ax.text(x + width/2, y + height/2, f'{height:.1f}%', 
                horizontalalignment='center', 
                verticalalignment='center',
                fontweight='bold',
                fontsize=11,
                color='white')

#adjust legend
handles, labels = ax.get_legend_handles_labels()
ax.legend(reversed(handles), reversed(labels), title='Primary Conduct', loc='center left', bbox_to_anchor=(1, 0.5))

plt.tight_layout()
plt.savefig('plots/pa_breakdown_primaryconduct_locations.png', dpi=200, bbox_inches='tight')
plt.show()
```

```python
query = '''
SELECT 
    bd.borough AS borough,
    c_d.not_deprived AS census_not_deprived,
    c_e.white AS census_white_ethnic,
    c_h.no_heating AS census_no_heating,
    c_o.hetero AS census_heterosexual,
    c_o.gay_lesbian AS census_gay_lesbian,
    c_o.no_answer AS census_no_answer_orientation,
    c_p.population_density AS census_pop_density,
    c_q.entry_level AS census_low_qualification,
    c_q.high_qualification AS census_high_qualification,
    c_r.born_uk AS census_born_uk,
    c_s.student AS census_student,
    s_c.crime_rate AS mps_crime_rate,
    (SELECT COUNT(*) FROM use_of_force uof2 WHERE uof2.PrimaryConduct > 2 AND uof2.borough_id = uof.borough_id AND uof2.Date_time > '2021-04-01 00:00:00') / (SELECT COUNT(*) FROM use_of_force uof3 WHERE uof3.borough_id = uof.borough_id AND uof3.Date_time > '2021-04-01 00:00:00') AS aggressive_conduct_ratio,
    c_e.total AS census_population,
    AVG(GREATEST(
        COALESCE(t1.ranking_1, 0),
        COALESCE(t2.ranking_1, 0),
        COALESCE(t3.ranking_1, 0),
        COALESCE(t4.ranking_1, 0),
        COALESCE(t5.ranking_1, 0),
        COALESCE(t6.ranking_1, 0),
        COALESCE(t7.ranking_1, 0),
        COALESCE(t8.ranking_1, 0),
        COALESCE(t9.ranking_1, 0),
        COALESCE(t10.ranking_1, 0),
        COALESCE(t11.ranking_1, 0),
        COALESCE(t12.ranking_1, 0),
        COALESCE(t13.ranking_1, 0),
        COALESCE(t14.ranking_1, 0),
        COALESCE(t15.ranking_1, 0),
        COALESCE(t16.ranking_1, 0),
        COALESCE(t17.ranking_1, 0),
        COALESCE(t18.ranking_1, 0),
        COALESCE(t19.ranking_1, 0),
        COALESCE(t20.ranking_1, 0)
    )) AS max_tactic
FROM use_of_force uof
#joins
LEFT JOIN ( 
SELECT 
	ROUND(not_deprived/total, 5) as not_deprived,
	ROUND(deprived_1dim/total, 5) as deprived_1dim,
    ROUND(deprived_2dim/total, 5) as deprived_2dim,
    ROUND(deprived_3dim/total, 5) as deprived_3dim,
    ROUND(deprived_4dim/total, 5) as deprived_4dim,
    borough_id
FROM census_deprivation
) c_d ON uof.borough_id = c_d.borough_id
LEFT JOIN ( 
SELECT 
    total as total,
	ROUND(white/total, 5) as white,
    borough_id
FROM census_ethnic
) c_e ON uof.borough_id = c_e.borough_id
LEFT JOIN ( 
SELECT
	ROUND(no_heating/total, 5) as no_heating,
    borough_id
FROM census_heating
) c_h ON uof.borough_id = c_h.borough_id
LEFT JOIN ( 
SELECT 
	ROUND(hetero/total, 5) as hetero,
    ROUND(gay_lesbian/total, 5) as gay_lesbian,
    ROUND(no_answer/total, 5) as no_answer,
    borough_id
from census_orientation
) c_o ON uof.borough_id = c_o.borough_id
LEFT JOIN ( 
SELECT 
	population_density,
    borough_id
from census_popdensity
) c_p ON uof.borough_id = c_p.borough_id
LEFT JOIN ( 
SELECT 
	ROUND((no_qualifications+level_1)/total, 5) as entry_level,
    ROUND(other_qualifications/total, 5) as high_qualification,
    borough_id
from census_qualification c_q
) c_q ON uof.borough_id = c_q.borough_id
LEFT JOIN ( 
SELECT 
	ROUND(born_uk/total, 5) as born_uk,
    borough_id
from census_residence c_r
) c_r ON uof.borough_id = c_r.borough_id
LEFT JOIN ( 
SELECT 
	ROUND(student/total, 5) as student,
    borough_id
FROM census_student c_s
) c_s ON uof.borough_id = c_s.borough_id
LEFT JOIN ( 
    SELECT * FROM crime_rates
) s_c ON uof.borough_id = s_c.borough_id
LEFT JOIN borough_dict bd ON uof.borough_id = bd.id
LEFT JOIN tactic_dict t1 ON uof.Tactic_1 = t1.id
LEFT JOIN tactic_dict t2 ON uof.Tactic_2 = t2.id
LEFT JOIN tactic_dict t3 ON uof.Tactic_3 = t3.id
LEFT JOIN tactic_dict t4 ON uof.Tactic_4 = t4.id
LEFT JOIN tactic_dict t5 ON uof.Tactic_5 = t5.id
LEFT JOIN tactic_dict t6 ON uof.Tactic_6 = t6.id
LEFT JOIN tactic_dict t7 ON uof.Tactic_7 = t7.id
LEFT JOIN tactic_dict t8 ON uof.Tactic_8 = t8.id
LEFT JOIN tactic_dict t9 ON uof.Tactic_9 = t9.id
LEFT JOIN tactic_dict t10 ON uof.Tactic_10 = t10.id
LEFT JOIN tactic_dict t11 ON uof.Tactic_11 = t11.id
LEFT JOIN tactic_dict t12 ON uof.Tactic_12 = t12.id
LEFT JOIN tactic_dict t13 ON uof.Tactic_13 = t13.id
LEFT JOIN tactic_dict t14 ON uof.Tactic_14 = t14.id
LEFT JOIN tactic_dict t15 ON uof.Tactic_15 = t15.id
LEFT JOIN tactic_dict t16 ON uof.Tactic_16 = t16.id
LEFT JOIN tactic_dict t17 ON uof.Tactic_17 = t17.id
LEFT JOIN tactic_dict t18 ON uof.Tactic_18 = t18.id
LEFT JOIN tactic_dict t19 ON uof.Tactic_19 = t19.id
LEFT JOIN tactic_dict t20 ON uof.Tactic_20 = t20.id
# Filters
WHERE uof.Outcome_Made_offescaped != 1 AND
uof.Incident_Location_Police_vehicle_with_prisoner_handling_cage != 1 AND
uof.Incident_Location_Police_vehicle_without_prisoner_handling_cage != 1 AND
uof.Incident_Location_Police_station_excluding_custody_block != 1 AND
uof.Incident_Location_Custody_Block != 1 AND
uof.Incident_Location_Ambulance != 1 AND
uof.Incident_Location_Sports_or_Event_Stadia != 1
GROUP BY 
    uof.borough_id,
    c_d.not_deprived,
    c_e.white,
    c_e.total,
    c_h.no_heating,
    c_o.hetero,
    c_o.gay_lesbian,
    c_o.no_answer,
    c_p.population_density,
    c_q.entry_level,
    c_q.high_qualification,
    c_r.born_uk,
    c_s.student,
    s_c.crime_rate;
'''
df = pd.read_sql_query(query, db_connection)
```

```python
#scale the population and population density variables
scaler = MinMaxScaler()
df[['census_population', 'census_pop_density']] = scaler.fit_transform(df[['census_population', 'census_pop_density']])

#bin the boroughs by the average of the measures
avg_measure = df.groupby('borough')['aggressive_conduct_ratio'].mean()
main_borough_bins = pd.qcut(avg_measure.values, q=2)
borough_to_group = {borough: "Borough_" + str(group) for borough, group in zip(avg_measure.index, main_borough_bins.codes)}
df['bg_aggr_ratio'] = df['borough'].map(borough_to_group)

avg_measure = df.groupby('borough')['mps_crime_rate'].mean()
main_borough_bins = pd.qcut(avg_measure.values, q=2)
borough_to_group = {borough: "Borough_" + str(group) for borough, group in zip(avg_measure.index, main_borough_bins.codes)}
df['bg_crime_rate'] = df['borough'].map(borough_to_group)

avg_measure = df.groupby('borough')['max_tactic'].mean()
main_borough_bins = pd.qcut(avg_measure.values, q=2)
borough_to_group = {borough: "Borough_" + str(group) for borough, group in zip(avg_measure.index, main_borough_bins.codes)}
df['bg_max_tactic'] = df['borough'].map(borough_to_group)

measures = ['census_pop_density', 'census_not_deprived', 'census_white_ethnic', 'census_high_qualification', 'census_student', 'census_born_uk']
borough_groups = ['bg_aggr_ratio', 'bg_crime_rate', 'bg_max_tactic']

plt.rcParams['font.size'] = 10
#plot the demographic measures for each borough bin
fig, axes = plt.subplots(nrows=1, ncols=6, figsize=(15, 5), dpi=200)

#mapping for the legend and colours
legend_mapping = {
    'Borough_0': 'Below median',
    'Borough_1': 'Above median'
}
color_mapping = {
    'Borough_0': 'blue',
    'Borough_1': 'red'
}

#plotting the bars
for i, measure in enumerate(measures):
    #get the axis
    ax = axes[i]
    #calculating means for each demographic measure across borough bins
    means = []
    for borough_group in borough_groups:
        group_means = df.groupby(borough_group)[measure].mean()
        means.append(group_means)
    #get the x-axis values
    ind = range(len(borough_groups)) 
    width = 0.35
    #plot the bars for borough groups
    for j, borough in enumerate(['Borough_0', 'Borough_1']):
        borough_means = [group_means[borough] for group_means in means]
        ax.bar([x + j*width for x in ind], borough_means, width=width, label=legend_mapping[borough], color=color_mapping[borough], alpha=0.6)
    ax.set_title(measure, fontsize=12)
    ax.set_xticks([x + 0.5*width for x in ind])
    ax.set_xticklabels([group_means.index.name for group_means in means])
    ax.legend()
    ax.set_ylim(0, 0.8)
    ax.tick_params(axis='x', labelsize=6)

plt.tight_layout()
plt.savefig('plots/pa_demographics_borough_bins.png', dpi=200, bbox_inches='tight')
plt.show()
```
